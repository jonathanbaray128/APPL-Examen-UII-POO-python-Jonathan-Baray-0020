```python
# Definición de la clase Empleado
class Empleado:
    # Método constructor para inicializar los 3 atributos
    def __init__(self, nombre, puesto, salario):
        self.nombre = nombre
        self.puesto = puesto
        self.salario = salario

    # Función 1: Leer/Confirmar los datos (Simulación de entrada)
    def leer_datos(self):
        print(f"--- Procesando información de: {self.nombre} ---")
        print(f"Estado: Datos del perfil '{self.puesto}' cargados correctamente.")

    # Función 2: Mostrar los datos del empleado
    def mostrar_datos(self):
        print("Ficha del Empleado:")
        print(f"  - Nombre: {self.nombre}")
        print(f"  - Puesto: {self.puesto}")
        print(f"  - Salario: ${self.salario}")
        print("-" * 30)

# --- Creación de Objetos (Instancias) ---

# Objeto empleado1
empleado1 = Empleado("", "", "")

# Objeto empleado2
empleado2 = Empleado("", "", "")

# --- Ejecución de las funciones ---

# Acciones para empleado1
empleado1.leer_datos()
empleado1.mostrar_datos()

# Acciones para empleado2
empleado2.leer_datos()
empleado2.mostrar_datos()
```
